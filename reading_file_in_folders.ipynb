```python
# import the modules
import os
from os import listdir
root = '../Images'
# get the path/directory
folder_dir = "/content/gdrive/MyDrive/Images"
imgs = os.listdir(folder_dir)
print(f'total number of images are {len(imgs)}')
for images in os.listdir(folder_dir):
    # check if the image ends with png
    if (images.endswith(".png") or images.endswith(".jpg")\
        or images.endswith(".jpeg")):
      f_path = os.path.join(root,images)
      print(f_path)
      im = cv.imread(f_path,cv.IMREAD_GRAYSCALE)
      plt.imshow(im)
```
