# Transformations using OpenCV

## Scaling

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
img = cv.imread('/content/lake.jpg')
```

```python
def scale(img,w,h):
  resized = cv.resize(img,(w,h))
  plt.imshow(resized)
  plt.show()
```

```python
scale(img,600,300)
```

## Translation

```python
def translate(img,a,b):
  height,width= img.shape[:2]

  trans_mtrx = np.float32([[1,0,a],[0,1,b]])
  rotated_image = cv.warpAffine(img, trans_mtrx, (width, height))

  plt.imshow(rotated_image)
```

```python
translate(img,100,50)
```

## Rotation

```python
def rotate(img,angl):
  height,width= img.shape[:2]

  center = (width/2, height/2)
  rotate_matrix = cv.getRotationMatrix2D(center=center, angle=angl, scale=1)

  rotated_image = cv.warpAffine(
    src=img, M=rotate_matrix, dsize=(width, height))

  plt.imshow(rotated_image)
```

```python
rotate(img,180)
```

## Affine Transformation

```python
def affine(img,pt1,pt2):
  pts1 = pt1
  pts2 = pt2
  rows,cols,ch = img.shape
        # Apply affine Transform Algorithm
  M = cv.getAffineTransform(pts1,pts2)
  dst = cv.warpAffine(img,M,(cols,rows))
  plt.imshow(dst)
```

```python
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
affine(img,pts1,pts2)
```

## Perspective Transformation

```python
def perspective(img,pt1,pt2):
      # Locate points of the documents
    # or object which you want to transform
  pts1 = pt1
  pts2 = pt2
        # Apply Perspective Transform Algorithm
  matrix = cv.getPerspectiveTransform(pts1, pts2)
  result = cv.warpPerspective(img, matrix, (500, 600))

  plt.imshow(result)
```

```python
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
perspective(img,pts1,pts2)
```

## Geometric Transformation 

```python
import cv2
import numpy as np

def geometric(img):
  num_rows, num_cols = img.shape[:2]
  rotation_matrix = cv2.getRotationMatrix2D((num_cols/2, num_rows/2), 30, 1)
  img_rotation = cv2.warpAffine(img, rotation_matrix, (num_cols, num_rows))
  plt.axis('off')
  plt.imshow(img_rotation)
  plt.show()
```

```python
geometric(img)
```
